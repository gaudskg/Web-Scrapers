```python
import requests
import string 
from bs4 import BeautifulSoup
import re
```

```python
key=[]
Enter_input = input("Search Country: ")
u_i=string.capwords(Enter_input)
lists=u_i.split()
word="_".join(lists)
url = "https://en.wikipedia.org/wiki/"+word
def wiki_scrape(url):
    print(url)
    url_open=requests.get(url)
    soup= BeautifulSoup(url_open.content,'html.parser')
    details= soup('table',{'class':'infobox'})
    for i in details:
        gdp_list=[]
        data=[]
        population=[]
        flag_link=[]
        payload_data = {}
        h = i.find_all('tr')
        for j in h:
            heading = j.find_all('th')
            detail = j.find_all('td')
            if heading is not None and detail is not None:   
                for x,y in zip(heading,detail):
                    if "Capital" in x.text:
                        payload_data['Capital'] = y.text
#                         print("Capital : ",y.text)
                    if "\xa0Total" in x.text:
#                         print("Total :",y.text)
                        data.append(y.text)
    
                    elif "\xa0census" in x.text or "estimate" in x.text:
                        pop = re.sub("[\(\[].*?[\)\]]", "", y.text)
                        population.append(pop)
                    
                    elif "Largest city" in x.text:
                            large_city = re.sub("[\(\[].*?[\)\]]", "", y.text)
                            large_city = large_city.strip()
                            payload_data['Largest City'] = large_city
#                             print("Largest City : ",large_city)
                            
                    elif "Official\xa0languages"==x.text:
                        language = re.sub("[\(\[].*?[\)\]\n]", "", y.text)
                        payload_data['Oficial Language'] = language
#                         print("Official Langauge : ",language)
                        
        for tab in soup.findAll('table',{'class':'infobox'}):
            flag = tab.find('img', attrs={'class':'thumbborder'})
            try:
                flag_link.append(flag.get("src"))
            except:
                pass
        
        
        
        if len(data)>=3 and len(population)>=1:
            payload_data['Total Area'] = data[0]
            payload_data['GDP Nominal'] = data[2]
            payload_data['Population'] = population[1]
            payload_data['Flag link'] = flag_link[0]
            print(payload_data)
#             print("Total Area : ",data[0])
#             print("GDP Nominal : ",data[2])
#             print("Population :",population[0])
#             print("Flag link : ",flag_link[0])
            
#             print()
wiki_scrape(url)
```

```python
country_list = ['China', 'United States', 'Japan', 'France', 'South Korea', 'United Kingdom', 'Canada', 'Germany', 'Spain', 'Australia', 'Brazil', 'Netherlands', 'Singapore', 
'Sweden', 'Italy', 'Switzerland', 'Austria', 'Belgium', 'Denmark', 'Finland', 'India', 'Luxembourg', 'Norway', 'Russia']
for i in country_list:
    u_i=string.capwords(i)
    lists=u_i.split()
    word="_".join(lists)
    url = "https://en.wikipedia.org/wiki/"+word
    print("********************************")
    wiki_scrape(url)
```
